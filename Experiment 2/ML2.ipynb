## 2. Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing, load_diabetes
```

```python
california = fetch_california_housing()
X_california = california.data[:, 3]
y_california = california.target 

X_california_mean = np.mean(X_california)
y_california_mean = np.mean(y_california)

# Compute slope (m) and intercept (b) for California
numerator_california = np.sum((X_california - X_california_mean) * (y_california - y_california_mean))
denominator_california = np.sum((X_california - X_california_mean) ** 2)
m_california = numerator_california / denominator_california
b_california = y_california_mean - m_california * X_california_mean

# Prediction for California
y_california_pred = m_california * X_california + b_california

# Plotting for California Housing
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_california, y_california, color='blue', label='Actual data')
plt.plot(X_california, y_california_pred, color='red', label='Regression line')
plt.xlabel('Average Rooms')
plt.ylabel('House Value')
plt.title('Linear Regression (California Housing Dataset)')
plt.legend()
plt.show()
print(f"Mean Squared Error (California Housing): {mse_california}")
```

```python
diabetes = load_diabetes()
X_diabetes = diabetes.data[:, 2]
y_diabetes = diabetes.target

X_diabetes_mean = np.mean(X_diabetes)
y_diabetes_mean = np.mean(y_diabetes)

numerator_diabetes = np.sum((X_diabetes - X_diabetes_mean) * (y_diabetes - y_diabetes_mean))
denominator_diabetes = np.sum((X_diabetes - X_diabetes_mean) ** 2)
m_diabetes = numerator_diabetes / denominator_diabetes
b_diabetes = y_diabetes_mean - m_diabetes * X_diabetes_mean

y_diabetes_pred = m_diabetes * X_diabetes + b_diabetes

plt.subplot(1, 2, 2)
plt.scatter(X_diabetes, y_diabetes, color='green', label='Actual data')
plt.plot(X_diabetes, y_diabetes_pred, color='red', label='Regression line')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression (Diabetes Dataset)')
plt.legend()

plt.tight_layout()
plt.show()

mse_california = np.mean((y_california - y_california_pred) ** 2)
mse_diabetes = np.mean((y_diabetes - y_diabetes_pred) ** 2)

print(f"Mean Squared Error (Diabetes): {mse_diabetes}")
```
