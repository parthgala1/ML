```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['species'])
print(one_hot_encoded_data)
```

```python
x = one_hot_encoded_data.iloc[:, :4].values
y = one_hot_encoded_data.iloc[:, 4:].values
```

```python
import matplotlib.pyplot as plt
color = ['red', 'blue', 'yellow', 'green', 'pink']
for i, column in enumerate(data.columns):
    plt.plot(data[column], color=color[i])
    plt.xlabel(column)
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
x_new = scale.fit_transform(x)
```

```python
x_new
```

```python
Q1 = np.percentile(one_hot_encoded_data, 25, axis=0)
Q3 = np.percentile(one_hot_encoded_data, 75, axis=0)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows where any feature is outside the bounds
data_no_outliers = one_hot_encoded_data[~((one_hot_encoded_data < lower_bound) | (one_hot_encoded_data > upper_bound)).any(axis=1)]
```

```python
data_no_outliers
```

```python
x = data_no_outliers.iloc[:, :4].values
y = data_no_outliers.iloc[:, 4:].values
```

```python
x_new = scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size=0.2, random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
x_test
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, np.argmax(y_train, axis=1))
```

```python
mnb.score(x_test, np.argmax(y_test, axis=1))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix
```

```python
y_pred.round().astype("int64")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert one-hot encoded to class labels
y_test_labels = np.argmax(y_test, axis=1)
y_pred_labels = np.argmax(y_pred, axis=1)

# Compute confusion matrix
cm = confusion_matrix(y_test_labels, y_pred_labels)

# Print confusion matrix
print("Confusion Matrix:")
print(cm)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], 
            yticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
